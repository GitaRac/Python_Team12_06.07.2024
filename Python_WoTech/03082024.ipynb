```python
# VALUE COUNTS!!! Create a grouped DataFrame that counts the number of Pokemons in each group for Type 1.

type1_pokemons = pokemon_df['Type 1'].value_counts()
print(type1_pokemons)
```

```python
#Bar chart

type1_pokemons.plot(x = 'Type1', y= 'count', kind = 'bar', color = 'white', edgecolor = 'black')
plt.title('Pokemon count by Type 1')
plt.xlabel('Type 1')
plt.ylabel('Count of Pokemons')
```

```python
#Which pokemon has the biggest attack?

pokemon_df['Attack'].max() #the highest value of the attack
pokemon_df['Attack'].idxmax() #will give you the index of the highest value

highest_attack = pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmax()] #iloc- index location
print('Pokemon with the highest Attack is:', highest_attack)
```

```python
#Which pokemon has the lowest attack?

pokemon_df['Attack'].min() #the lowest value of the attack
pokemon_df['Attack'].idxmin() #will give you the index of the lowest value

lowest_attack = pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmin()] #iloc- index location
print('Pokemon with the lowest Attack is:', lowest_attack)
```

##HISTOGRAM

```python
#Analyze Total for each pokemon
plt.hist(pokemon_df['Total'], color='green', edgecolor='black', alpha=0.7)

plt.xlabel('Total')
plt.ylabel('Number of pokemons(Frequency)')

#avg line
plt.axvline(pokemon_df['Total'].mean(), linestyle='dashdot', color='red')

#min line
plt.axvline(pokemon_df['Total'].min(), linestyle='dashdot', color='pink')

#max line
plt.axvline(pokemon_df['Total'].max(), linestyle='dashdot', color='green')

plt.grid(linestyle='--', alpha=0.7)
```

```python
value = plt.hist(pokemon_df['Total'], bins = 3, color='green', edgecolor='black', alpha=0.7)

hist_values = value[0]
bin_edges = value[1]

print('Histogram counts:', hist_values)
print('Bin edges', bin_edges)
```

### 1. How many Pokémons are with 'Type 1' == Water as a % of total?

```python
water_pokemon_count = (pokemon_df['Type 1'] == 'Water').sum()
percentage = water_pokemon_count / len(pokemon_df) * 100

print(f"The percentage of Pokémon with 'Type 1' as Water is {percentage:.1f}%")
```

### 2. What is the maximum 'Speed' value? What is the minimum 'Speed' value? What is the difference between max and min 'Speed'?

```python
max_speed = pokemon_df['Speed'].max()
min_speed = pokemon_df['Speed'].min()
speed_difference = max_speed - min_speed

print(f'The maximum Speed value is {max_speed}. Minimum Speed value is {min_speed}. The difference between max and min Speed is {speed_difference}')
```

### 3. Filter the DataFrame to include only the Pokémon with 'Speed' >= 80. How many Pokémon meet this criterion? Display this DataFrame using your preferred visualization method.

```python
speed_greater_than_80 = pokemon_df[pokemon_df['Speed'] >= 80]
print(f'Number of Pokemons with Speed equal or greater than is: {len(speed_greater_than_80)}')

plt.hist(pokemon_df['Speed'], bins=15, color='lavenderblush', edgecolor='maroon')
plt.xlabel('Speed')
plt.ylabel('Number of pokemons (Frequency)')
plt.axvline(80, linestyle='dashdot', color='maroon')
```

### 4. (DIFFICULT) Find Pokémon with the longest name (excluding spaces)? What is this pokemons name?

```python
# Find the Pokémon with the longest name by length of the name without spaces
longest_pokemon_name = pokemon_df['Name'].apply(lambda x: len(x.replace(" ", ""))).idxmax()

# Retrieve the name of the Pokémon with the longest name
longest_pokemon_name = pokemon_df.loc[longest_pokemon_name, 'Name']

print(f'The longest Pokémon name is {longest_pokemon_name}.')
```
